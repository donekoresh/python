## shallow copy

```python
x=[1,2,3,4]
y=x
print(x)
print(y)
y[1]=88
print(x)
print(y)
```

## deep copy

```python
x=[1,2,3,4]
y=x.copy()
print(x)
print(y)
y[1]=88
print(x)
print(y)
```

```python
x=[1,2,3,4]
y=x[:]
print(x)
print(y)
y[1]=88
print(x)
print(y)
```
